```python
# Import the required python libraries

import pandas as pd
import numpy as np

# Import the dataset

file = "https://raw.githubusercontent.com/pranithgunda/HF-DA-Projects/main/MER_T12_06.csv"
df = pd.read_csv(file)

# Read the Dataset

df.head(20)
```

```python
df.tail(20)
```

```python
# Dataset contains data for Carbon Emissions from various sources for period Jan 1973 to July 2016.
# Dataset contains observation for 13th month for a given year. These data records need to be cleaned up for observed yrs
```

```python
# Unique sources of carbon emissions available in the Dataset

df['Description'].unique()
```

```python
# Check Datatypes for columns available in the Dataset
df.info()
```

```python
# Lets create new column to extract the month value from column 'YYYYMM'

df['MM'] = df['YYYYMM'].astype(str).str[4:]
```

```python
# Read the Dataset for new column
df.head(20)
```

```python
# From the Dataset, drop the records having 'MM' value as 13

df.drop(df[df['MM'] == '13'].index, inplace = True)
```

```python
# Read the Dataset, to confirm the deletion of records

df.head(20)
```

```python
# Lets check the Datatype for all columns in the Dataset

df.info()
```

###### Dataset size is now reduced to 4707 records after removing the observations for 13th month

```python
# Lets convert the Datatype for column 'Value' as float

# df.astype({"Value":float})
```

###### Column "Value" has "Not Available" values from the execution of above cell, below error is observed

ValueError: could not convert string to float: 'Not Available': Error while type casting for column 'Value'

```python
# Query for 'Not Available' values

df.query("Value == 'Not Available'")
```

```python
# 384 missing values are observed for column 'Value'
```

```python
# 'Not Available' is not recognized as missing value in python.

df.isnull().sum()
```

```python
# Lets replace "Not Available" values with Standard Missing values for python to recognize

df = df.replace('Not Available',np.nan)
```

```python
# Check for Standard Missing values in the Dataset now

df.isnull().sum()
```

```python
# Standard missing values are now observed by the python
```

```python
# Remove the missing values from Dataset now

df = df.dropna()
```

```python
df.info()
```

###### Dataset size is now reduced to 4323 rows after removing the standard missing values

```python
# Now lets change the datatype for column 'Value' to float to perform aggregrations

df['Value'] = pd.to_numeric(df['Value'])
```

```python
# Check Datatypes for all columns now

df.info()
```

```python
# Lets extract 'Year' value from column 'YYYYMM' into sepearte column

df['YYYY'] = df['YYYYMM'].astype(str).str[:4]
```

```python
# Check the new column 'YYYY'
df.head()
```

```python
# Check the Datatype for new column 'YYYY'

df.info()
```

```python
# Rename column name 'Descriptio' to 'Source'

df = df.rename(columns={'Description':'Source'})
```

```python
# Read the Dataset to confirm column rename

df.head()
```

```python
# Remove unwanted columns from the Dataset

# Column 'Column_Order' has no significance

df = df.drop(['Column_Order'], axis = 1)
```

```python
# Read the Dataset to confirm column deletion

df.head()
```

```python
# Group By Year Wise to calculate the total amount of Carbon Emitted from various energy sources

df.groupby(['Source'])['Value'].sum().sort_values(ascending = False)
```

****###### 'Total Energy Electric Power Sector CO2 Emissions' has emitted highest amounts of Carbon for period Jan 1973 - July 2013

```python
# Lets create a monthly Dataframe

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

df_monthly = df[['Source','Value','YYYYMM','YYYY','MM']].copy()

df_monthly.head()
```

```python
# Create new dataset to calculate the mean value

df_Yearly = df_monthly.groupby(['Source','YYYYMM','YYYY','MM']).mean()
```

```python
df_Yearly = df_Yearly[['Value']].reset_index()
```

```python
df_Yearly.head()
```

```python
# Plot the Line Plot

sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(18,8)})

ax = sns.lineplot(data = df_Yearly, x = 'YYYY', y = 'Value', hue = 'Source', palette = 'viridis', legend = 'full', lw=4)
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
plt.legend(bbox_to_anchor = (1,1))
plt.ylabel('Carbon Emissions Per Energy Source (MMT)')
plt.xlabel('Year')
plt.show()
```

```python
# Create Dataframe for top 5 emissions

df_topemissions = df.loc[(df['Source'] == 'Total Energy Electric Power Sector CO2 Emissions') | (df['Source'] == 'Coal Electric Power Sector CO2 Emissions') | (df['Source'] == 'Natural Gas Electric Power Sector CO2 Emissions') | (df['Source'] == 'Petroleum Electric Power Sector CO2 Emissions') | (df['Source'] == 'Residual Fuel Oil Electric Power Sector CO2 Emissions') ]

df_topemissions['Source'].unique()
```

```python
# Read the columns copied

df_topemissions.columns
```

```python
# Drop unwanted columns

df_topemissions = df_topemissions.drop(['MSN','YYYYMM','Unit','MM','YYYY'], axis = 1)
```

```python
# Read the Dataframe to confirm the columns

df_topemissions.head()
```

```python
# Plot pie chart for top 5 emissions

# Group the data by 'Source' and calculate the sum
grouped_data = df_topemissions.groupby(['Source']).sum()

# Calculate the percentage for each source
total_emissions = grouped_data['Value'].sum()
grouped_data['Percentage'] = grouped_data['Value'] / total_emissions * 100

# Plot the pie chart with percentages
plt.figure(figsize=(8, 8))
grouped_data['Value'].plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.legend(grouped_data.index, loc='center left', bbox_to_anchor=(1, 0.5))

# Display the plot
plt.show()

```

Total Energy Electric Power Sector CO2 Emissions has highest amount of emission with 48.8% among top 5 sources.

Conclusion: Individual Trend Analysis can be observed for all sources of Carbon Emission from Line Plot Graph. 
Carbon Emissions from Coal Electic Power Sector and Total Energy Electric Power Sector showed an increased trend 
from 1975 till 2007. However Carbon Emissions from Natural Gas Electric Power Sector showed an increased trend 
from 1988, which is a good sign as electricity generation from Natural Gases are efficient compared to other sources
of electricity generation.

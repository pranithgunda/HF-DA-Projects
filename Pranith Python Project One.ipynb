```python
#importing necessary packages for our project
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
url="https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv"
```

```python
# Read dataset
df=pd.read_csv(url)
```

```python
df
```

```python
# Replace "na" and "S" values to 0, as part of cleaning task
dfs=df.replace({'na':'0','s':'0'})
```

```python
# Removing percent columns, to utilize columns with Number
dfs=dfs.drop(['Percent Scoring Below 65','Percent Scoring 65 or Above','Percent Scoring 80 or Above','Percent Scoring CR'], axis=1)
```

```python
dfs
```

```python
dfs.dtypes
```

```python
# Convert data types to integer
for i in dfs.columns:
    try:
        dfs[[i]] = dfs[[i]].astype(float).astype(int)
    except:
        pass

print(dfs.dtypes)
```

```python
# Confirm Data Types after conversion
dfs.dtypes
```

```python
# Count the school names
dfs['School Name'].value_counts()
```

```python
# Count different Regents Exam
dfs['Regents Exam'].value_counts()
```

```python
# Group By School Name and Regents Exam to calculate sum of Mean Score
dfs.groupby(['School Name', 'Regents Exam'])['Mean Score'].sum()
```

```python
# Group By School Name and Regents Exam to calculate counts
dfs.groupby(['School Name'])['Regents Exam'].value_counts()[:10]
```

```python
#there are total of 1012 unique schools
dfs['School Name'].nunique() 
```

```python
#checking if there are any null values in the dataset
for col in dfs.columns:
    count=dfs[col].isna().sum()
    print(col,count)
```

```python
#Descriptive statistics
dfs.describe()
```

```python
# Display Top 10 school names with sum of Total Tested
s=dfs.groupby(['School Name'])['Total Tested'].sum().sort_values(ascending=False)[:10]
```

```python
sa=pd.DataFrame(s)
```

```python
sa
```

```python
#Total school levels in each concentration and High School has total 151160.
dfs['School Level'].value_counts()
```

```python
#Showing the top schools with total tested 
df2 = dfs.groupby(['School Name','Regents Exam','Mean Score'])['Total Tested'].sum().sort_values(ascending=False)
```

```python
df2
```

```python
#Creating a dataframe
data_file=pd.DataFrame(df2)
```

```python
frames=data_file.reset_index()
```

```python
frames
```

```python
#Converting to int
for i in frames.columns:
    try:
        frames[[i]] = frames[[i]].astype(float).astype(int)
    except:
        pass

print(frames.dtypes)
```

```python
#Data Distribution of total tested  feature.
sns.distplot(dfs['Total Tested'])
```

```python
#Bar-chart visualizations of Regent Exams
plt.figure(figsize=(28,5))
sns.countplot(data=dfs,y='Regents Exam', orient='h')
plt.yticks(np.arange(5), ['Common Core Algebra','Living Environment','U.S. History and Government','Common Core English','Global History and Geography'])
plt.ylabel('Regents Exam')
plt.xlabel(None)
plt.show()
```

```python
#Feature relationship between Mean Score and Total Testes
frames.plot.scatter("Mean Score", "Total Tested")
plt.show()
```
